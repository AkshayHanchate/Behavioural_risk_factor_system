```python
# Remove the 'b' and quotes from the byte string columns
def clean_byte_string(byte_str):
    if isinstance(byte_str, bytes):
        return byte_str.decode('utf-8')
    return byte_str

# Apply the cleaning function to the relevant columns
df['IMONTH'] = df['IMONTH'].apply(clean_byte_string)
df['IYEAR'] = df['IYEAR'].apply(clean_byte_string)
df['IDAY'] = df['IDAY'].apply(clean_byte_string)
df['IDATE'] = df['IDATE'].apply(clean_byte_string)

# Convert the cleaned columns to integers (if necessary)
df['IMONTH'] = pd.to_numeric(df['IMONTH'], errors='coerce')
df['IYEAR'] = pd.to_numeric(df['IYEAR'], errors='coerce')
df['IDAY'] = pd.to_numeric(df['IDAY'], errors='coerce')
df['IDATE'] = pd.to_numeric(df['IDATE'], errors='coerce')
df['SEQNO'] = pd.to_numeric(df['SEQNO'], errors='coerce')

# Display the cleaned DataFrame
print(df[['IMONTH', 'IYEAR','IDAY','IDATE']].head())
```
